```python
#Python Week 2 - Day 2 Tic-Tac-Toe Homework

num = input("Enter the number")
num = int(num)

hpos=num
tot =  (3 * num)+2 #total length
totn=tot

vpos=num
vern=2

v1=3 #No. of iterations
ct=vern
str1 = ""
f = open("demofile.txt", "w")

while v1>0:
    while vpos>0:
        while ct>0:
            while hpos>0:
                str1 = str1 + " "
                hpos=hpos-1
                                
            str1=str1+ "|"
            hpos=num
            ct = ct-1   
        f.write(str1)
        str1="\n"
        vpos=vpos-1
        ct=vern
    if v1 > 1:      
        while totn>0:
            str1=str1+"-"
            totn=totn-1
        f.write(str1)
        str1="\n" 
        totn=tot 
    v1=v1-1
    vpos=num
      
f.close() 
```
