```python
#Python Week 2 - Day 2 Tic-Tac-Toe Homework

num = input("Enter the number")
num = int(num)

hpos=num
tot =  (3 * num)+2 #total length
totn=tot

vpos=num
vern=2

v1=3 #No. of iterations
ct=vern
str1 = ""
f = open("demofile.txt", "w")

while v1>0:
    while vpos>0:
        while ct>0:
            while hpos>0:
                str1 = str1 + " "
                hpos=hpos-1
                                
            str1=str1+ "|"
            hpos=num
            ct = ct-1   
        f.write(str1)
        str1="\n"
        vpos=vpos-1
        ct=vern
    if v1 > 1:      
        while totn>0:
            str1=str1+"-"
            totn=totn-1
        f.write(str1)
        str1="\n" 
        totn=tot 
    v1=v1-1
    vpos=num
      
f.close() 
```

```python
# Home work 3

import secrets
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes


key = secrets.token_bytes(32)
f = open("encry_key.bin", "wb")
f.write(key)

plain_text = input("Enter the string to encrypt : ")



def encrypt_string111(str_val):
    
    f = open("encry_key.bin", "rb")
    f.read()
    
    # generating a random number

    nonce= get_random_bytes(16)
    # Create a cipher object for decryption
    cipher = AES.new(key, AES.MODE_EAX,nonce=nonce)

    cipher_text, tag = cipher.encrypt_and_digest(str_val.encode('utf-8'))
    encrypted_text = nonce + cipher_text + tag
    return encrypted_text

def decrypt_string111(ciph_text):
   
    # Extracting nonce, cipher text & tag from input
    nonce = ciph_text[:16]
    ciphertext = ciph_text[16:-16]
    
    tag = ciph_text[-16:]

    # Create a cipher object for decryption
    
    cipher = AES.new(key, AES.MODE_EAX,nonce=nonce)
    decrypted_text = cipher.decrypt_and_verify(ciphertext,tag)
    

    return decrypted_text.decode('utf-8')
        

ec_text = encrypt_string111(plain_text)
print (f"Encrypted text : {ec_text[16:-16]}")

dc_text = decrypt_string111(ec_text)
print (f"Decrypted text : {dc_text}")
```

```python
#Week 2 Day 4 - Home work 4 
# Taken from ChatGPT

from Crypto.PublicKey import RSA
from Crypto.Signature import pkcs1_15
from Crypto.Hash import SHA256

# Generate RSA key pair
key = RSA.generate(2048)

private_key = key.export_key()
public_key = key.publickey().export_key()

# Message to be signed
input_message = input("Enter the message to sign : ")
input_message = input_message.encode('utf-8')

# Create SHA-256 hash of the message
hash_object = SHA256.new(input_message)

# Load private key for signing
private_key = RSA.import_key(private_key)

# Sign the hash using the private key
signature = pkcs1_15.new(private_key).sign(hash_object)

# Print the signature
print("Signature:", signature)

# Load public key for verification
public_key = RSA.import_key(public_key)

# Verify the signature
try:
    pkcs1_15.new(public_key).verify(hash_object, signature)
    print("Signature is valid.")
except (ValueError, TypeError):
    print("Signature is invalid.")
```

```python
#Week 2 Day 5- Home work-6

import math

class Shape:

    def area(self):
        pass

class Triangle(Shape):
    def __init__(self):
        self.a = int(input("Enter length of 1st side of triangle"))
        self.b = int(input("Enter length of 2nd side of triangle"))
        self.c = int(input("Enter length of 3rd side of triangle"))

        assert (self.a+self.b) > self.c, "a+b < c : So cannnot form a triangle"
        assert (self.a+self.c) > self.b, "a+c < b : So cannnot form a triangle"
        assert (self.c+self.b) > self.a, "b+c < a : So cannnot form a triangle"
     
    def area(self):
        s=(self.a+self.b+self.c)/2
        area_triangle= math.sqrt(s*(s-self.a)*(s-self.b)*(s-self.c))
        print(f"Area of triangle with sides {self.a}, {self.b} & {self.c} : {area_triangle}")


class Circle(Shape):
    def __init__(self):
        self.r = int(input("Enter the radius of length of circle"))

    def area(self):
        area_circle = math.pi * self.r * self.r
        print(f"\nArea of circle with radius {self.r} : {area_circle}")   
        

class Rectangle(Shape):
    def __init__(self):
        self.a = int(input("Enter length of rectangle"))
        self.b = int(input("Enter breadth of rectangle"))
            
    def area(self):
        area_rectangle = self.a * self.b
        print(f"\nArea of rectangle with length of {self.a} & breadth of {self.b} : {area_rectangle}")  

class Square(Shape):
    def __init__(self):
        self.a = int(input("Enter length of a side"))    

    def area(self):
        area_square = self.a * self.a
        print(f"\nArea of square of side length of {self.a} : {area_square}")   


my_shape = Triangle() 
my_shape.area()     

my_shape = Circle()
my_shape.area()

my_shape = Rectangle()
my_shape.area()

my_shape = Square()
my_shape.area()

```
